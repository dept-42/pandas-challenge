## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary for later aggregations
school_spending_df = per_school_summary_unformated.copy()

school_spending_df.dtypes
```

```python
school_spending_df
```

```python
# Use `pd.cut` on the per_school_capita Series from earlier to categorize per student spending based on the bins.
#school_spending_df['Per Student Budget']= per_school_capita['Per Student Budget'].astype(int)
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(x=per_school_capita, bins=spending_bins, labels=labels)
# Convert Spending Ranges (Per Student) to a string
school_spending_df["Spending Ranges (Per Student)"]= school_spending_df["Spending Ranges (Per Student)"].astype('string')
school_spending_df.dtypes
#school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Math Score Avg"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Reading Score Avg"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Math Pass Rate"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Reading Pass Rate"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Pass Rate"].mean()
spending_math_scores
```

```python
# Assemble into DataFrame
#convert to lists
spending_math_scores = spending_math_scores.values.tolist()
spending_reading_scores = spending_reading_scores.values.tolist()
spending_passing_math = spending_passing_math.values.tolist()
spending_passing_reading = spending_passing_reading.values.tolist()
overall_passing_spending = overall_passing_spending.values.tolist()

#assemble dictionary
spending_summary_dict = {
    "Spend by Average Math Score": spending_math_scores,
    "Spend by Reading Score": spending_reading_scores,
    "Spend by Math %Pass": spending_passing_math,
    "Spend by Reading %Pass": spending_passing_reading,
    "Spend by Overall %Pass": overall_passing_spending}

#Convert to DataFrame
spending_summary=pd.DataFrame(spending_summary_dict, index=[labels])

#format
spending_summary["Spend by Average Math Score"] = spending_summary["Spend by Average Math Score"].map("${:.2f}".format)
spending_summary["Spend by Reading Score"] = spending_summary["Spend by Reading Score"].map("${:.2f}".format)
spending_summary["Spend by Math %Pass"] = spending_summary["Spend by Math %Pass"].map("${:.2f}".format)
spending_summary["Spend by Reading %Pass"] = spending_summary["Spend by Reading %Pass"].map("${:.2f}".format)
spending_summary["Spend by Overall %Pass"] = spending_summary["Spend by Overall %Pass"].map("${:.2f}".format)
# Display results
spending_summary
```
